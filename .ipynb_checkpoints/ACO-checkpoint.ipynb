```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ant_colony_optimization(graph, start_node, num_ants=10, num_iterations=100, evaporation_rate=0.5, alpha=1, beta=2):
    num_nodes = len(graph)
    pheromone_level = [[1.0] * num_nodes for _ in range(num_nodes)]
    best_path = None
    best_path_length = float("inf")

    for iteration in range(num_iterations):
        for ant in range(num_ants):
            current_node = start_node
            path = [current_node]
            visited_nodes = [False] * num_nodes
            visited_nodes[current_node] = True

            while len(path) < num_nodes:
                unvisited_nodes = [node for node in range(num_nodes) if not visited_nodes[node]]
                if not unvisited_nodes:
                    break

                # Calculate the probability of choosing each unvisited node
                probabilities = [0.0] * num_nodes
                total_probability = 0.0
                for node in unvisited_nodes:
                    distance_value = graph[current_node][node]
                    pheromone_value = pheromone_level[current_node][node]
                    probability = pow(pheromone_value, alpha) * pow(1.0 / distance_value, beta)
                    probabilities[node] = probability
                    total_probability += probability

                # Choose the next node based on the probabilities
                random_value = random.uniform(0, total_probability)
                next_node = None
                probability_sum = 0.0
                for node, probability in enumerate(probabilities):
                    probability_sum += probability
                    if probability_sum >= random_value:
                        next_node = node
                        break

                # Update the current node and path
                current_node = next_node
                path.append(current_node)
                visited_nodes[current_node] = True

            # Calculate the path length of the ant's path
            path_length = sum([graph[path[i]][path[i+1]] for i in range(num_nodes-1)])

            # Update the pheromone levels on the ant's path
            for i in range(num_nodes-1):
                node1, node2 = path[i], path[i+1]
                pheromone_level[node1][node2] = (1 - evaporation_rate) * pheromone_level[node1][node2] + evaporation_rate * (1.0 / path_length)
                pheromone_level[node2][node1] = (1 - evaporation_rate) * pheromone_level[node2][node1] + evaporation_rate * (1.0 / path_length)

            # Update the best path if the ant found a shorter path
            if path_length < best_path_length:
                best_path = path
                best_path_length = path_length

        print(f"Iteration {iteration}: Best path length = {best_path_length}")

    return best_path, best_path_length
```

```python
graph = [
    [0, 10, 15, 20],
    [10, 0, 38, 25],
    [15, 35, 0, 30],
    [20, 25, 40, 0]
]

start_node = 0
num_ants = 10
num_iterations = 100
evaporation_rate = 0.5
alpha = 1
beta = 2

best_path, best_path_length = ant_colony_optimization(
    graph, start_node, num_ants=num_ants, num_iterations=num_iterations,
    evaporation_rate=evaporation_rate, alpha=alpha, beta=beta
)

print(f"Best path: {best_path}")
print(f"Best path length: {best_path_length}")
```

```python
best_path = [0, 2, 1, 3] # example best path

# Plot the nodes
for i in range(len(graph)):
    plt.scatter(i+1, i+1)
    
# Plot the edges
for i in range(len(graph)):
    for j in range(i+1, len(graph)):
        plt.plot([i+1, j+1], [j+1, i+1], 'k-')

# Plot the best path
for i in range(len(best_path)-1):
    plt.plot([best_path[i]+1, best_path[i+1]+1], [best_path[i+1]+1, best_path[i]+1], 'r-', linewidth=2)

plt.show()
```
